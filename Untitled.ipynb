```python
image_sequence = []

for epoch in range(1, 21):
    image = media.read_image(os.path.join(args.result_path, f"{epoch}.png"))
    image_sequence.append(image.reshape(1, image.shape[0], image.shape[1], image.shape[2]))

gt = media.read_image(os.path.join(args.result_path, "GT.png"))    
gt = gt.reshape(1, gt.shape[0], gt.shape[1], gt.shape[2])
gt_sequence = [gt for _ in range(1, 21)]
image_sequence = np.concatenate(image_sequence, axis =0)[:, :, :, :3]
gt_sequence = np.concatenate(gt_sequence, axis =0)[:, :, :, :3]
sequence = np.concatenate([image_sequence, gt_sequence], axis=2)
```

```python
video = ImageSequenceClip([image for image in sequence], fps=3)

video_name = "learning process"
video.write_videofile(f"{video_name}.mp4") 
# media.show_video(image_sequence, fps=2)
# media.show_images(image_sequence, height=80, columns=4)
```
